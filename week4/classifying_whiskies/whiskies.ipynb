```python
import numpy as np
import pandas as pd
```

```python
# Read whiskies.txt to a whiky dataframe
whisky = pd.read_csv("whiskies.txt")
# Add Region column from regions.txt
whisky["Region"] = pd.read_csv("regions.txt")
# Get top 5
whisky.head()
```

```python
# Flavors dataframe
# Using iloc to get subset of whisky dataframe
flavors = whisky.iloc[:, 2:14]
flavors.head()
```

```python
# flavors correlation matrix
corr_flavors = pd.DataFrame.corr(flavors)
corr_flavors
```

```python
# vizualize corr_flavors
import matplotlib.pyplot as plt
# vizualize
plt.figure(figsize=(10,10))
plt.pcolor(corr_flavors)
plt.colorbar()
plt.show()
```

```python
# transposed corr_flavors
transposed_flavors = flavors.transpose()
transposed_flavors
```

```python
# transposed flavors correlation matrix
corr_trans_flavors = pd.DataFrame.corr(flavors.transpose())
corr_trans_flavors
```

```python
# vizualize transposed flavors
plt.figure(figsize=(10,10))
plt.pcolor(corr_trans_flavors)
plt.colorbar()
plt.show()
```
